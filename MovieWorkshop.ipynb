### Filtering Data in Pandas
Filtering is an essential part of data analysis, allowing you to focus on rows that match specific criteria. Below, we will explore different ways to filter data using `.loc[]` with various conditions.


### Filtering Rows by Specific Value
To filter rows based on a specific value in a column, we use the `==` operator. This is useful when you want to see all rows that contain a specific value in a certain column.


````python
# Example: Filter movies with Genre equal to "Action"
action_movies = df.loc[df['Genre'] == 'Action']
action_movies.head()
````

### Filtering Numeric Values with Comparison Operators
You can use comparison operators like `>`, `<`, `>=`, or `<=` to filter rows based on numerical conditions. For example, finding all movies with an IMDB rating greater than 8.


````python
# Example: Filter movies with IMDB Rating greater than 8
high_rated_movies = df.loc[df['IMDB_Rating'] > 8]
high_rated_movies.head()
````

### Filtering by Categorical Variable
When filtering categorical data, you can also specify text values directly. For example, selecting a particular movie title.


````python
# Example: Filter for a specific movie by title
specific_movie = df.loc[df['Series_Title'] == 'Inception']
specific_movie
````

### Filtering with Multiple Conditions
You can combine multiple filtering conditions using `&` (AND) and `|` (OR) to get more refined results.


````python
# Example: Filter movies with Genre 'Action' AND IMDB Rating greater than 8
action_and_high_rated = df.loc[(df['Genre'] == 'Action') & (df['IMDB_Rating'] > 8)]
action_and_high_rated.head()

# Example: Filter movies that are either in the Genre 'Comedy' OR have an IMDB Rating above 8.5
comedy_or_high_rated = df.loc[(df['Genre'] == 'Comedy') | (df['IMDB_Rating'] > 8.5)]
comedy_or_high_rated.head()
````

# Data Cleaning

````python
# Remove Columns: Keep only the columns Series_Title, Released_Year, Genre, IMDB_Rating, Runtime, and Gross
movies = movies[['Series_Title', 'Released_Year', 'Genre', 'IMDB_Rating', 'Runtime', 'Gross']]

# Modify Genre: Keep only the first value before the first comma
movies['Genre'] = movies['Genre'].str.split(',').str[0]

# Rename Gross Column to Revenue
movies.rename(columns={'Gross': 'Revenue'}, inplace=True)

# Clean Runtime: Remove " min" and convert to int
movies['Runtime'] = movies['Runtime'].str.replace(' min', '').astype(int)
````

````python
# First, identify and remove any non-numeric values in the "Released_Year" column
movies = movies[movies["Released_Year"].str.isnumeric()]

# Now, convert the "Released_Year" column to int
movies["Released_Year"] = movies["Released_Year"].astype(int)
````

````python
movies['Revenue'] = movies['Revenue'].str.replace(',', '')

# Step 2: Convert the 'Revenue' column to numeric
movies['Revenue'] = pd.to_numeric(movies['Revenue'], errors='coerce')

movies['Revenue'] = movies['Revenue'] / 1e6  # Now in millions

movies["Revenue"]
````

````python
movies.rename(columns= {"Series_Title":"Title","IMDB_Rating":"Rating","Released_Year":"Release"},inplace=True)
````

### Box Plot

A **box plot** (also called a **box-and-whisker plot**) is a graphical representation of the distribution of data that shows the median, quartiles, and potential outliers. Its especially useful for understanding the spread and skewness of the data.

#### Key Components:
- **Median (Q2)**: The line in the center of the box represents the median value of the dataset.
- **First Quartile (Q1)**: The lower edge of the box represents the 25th percentile.
- **Third Quartile (Q3)**: The upper edge of the box represents the 75th percentile.
- **Interquartile Range (IQR)**: The range between Q1 and Q3, representing the middle 50% of the data.
- **Whiskers**: Lines extending from the box that show the range of data within 1.5 times the IQR. Points beyond the whiskers are considered **outliers**.

Box plots are particularly useful for identifying **outliers** and comparing the distribution of data across different categories.

### Creating a Box Plot with Seaborn

In Python, you can create a box plot easily using **Seaborn's** `boxplot()` function. This function takes several important arguments:

- **`x`**: The categorical variable for the x-axis.
- **`y`**: The numerical variable for the y-axis.
- **`data`**: The dataset from which to draw the plot.



````python
sns.catplot(data=movies,x="Revenue",kind="box",col="Genre",col_wrap=3)
plt.show()
````

### Seaborn's `displot()`

**`displot()`** is a flexible function in Seaborn used to visualize the distribution of a dataset. It is a **figure-level function** that can create **histograms**, **kernel density plots**, or **ECDF plots** (Empirical Cumulative Distribution Functions). It provides more flexibility compared to `histplot()` by offering multiple types of distribution plots.

### Parameters
- **`data`**: The dataset from which to draw the plot.
- **`x`**: The variable to plot on the x-axis.
- **`kind`**: Specifies the type of plot to be created. It can take one of the following values:
  - **`"hist"`**: Creates a histogram (default).
  - **`"kde"`**: Creates a kernel density estimate plot.
  - **`"ecdf"`**: Creates an ECDF plot.



````python
sns.displot(data=movies,x="Runtime",col="Genre",col_wrap=3,kind="hist",bins=20)
plt.show()
````

### Grouping Data with `groupby()`

The **`groupby()`** method in pandas is used to group data based on the values in one or more columns. It allows for easier aggregation, summarization, and analysis of data by dividing the dataset into groups, making it easier to analyze subgroups.

The typical workflow for using `groupby()` is to **group** the data by one or more columns and then **apply aggregation functions** like `.sum()`, `.mean()`, `.size()`, etc.

#### Key Steps in Using `groupby()`:
1. **Grouping**: Split the dataset into groups based on unique values in the selected columns.
2. **Aggregation**: Apply a function like `mean`, `count`, or `sum` to calculate metrics for each group.

#### Syntax
```python
df.groupby("column_name").aggregation_function()


### Key Features of `groupby()`
- **Flexible Grouping**: You can group by one or multiple columns to gain insights at different levels.
- **Aggregation Functions**: You can use a wide range of functions such as `.sum()`, `.mean()`, `.count()`, `.max()`, `.min()`, etc., to summarize the grouped data.
- **Data Analysis**: `groupby()` is particularly useful for exploratory data analysis, enabling you to derive insights like category-wise averages, totals, counts, etc.


**Grouping the data by Genre and finding the Average rating for each Genre**

````python
AvgRating_ByGenre = movies.groupby("Genre")["Rating"].mean().reset_index(name="avg_Rating").round(2)
AvgRating_ByGenre
````

A **bar plot** is a type of chart that represents data with rectangular bars. The length of each bar corresponds to the value it represents. Bar plots are used to compare quantities across different categories, making it easy to visualize and compare summary statistics, like means or counts, for categorical data.

### Creating a Bar Plot with Seaborn

In Python, you can use **Seaborn's** `barplot()` function to create a bar plot. This function takes several important arguments:
